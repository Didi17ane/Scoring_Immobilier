```python
from graphviz import Digraph

# Création du diagramme
dot = Digraph(comment="ERD Tylimmo - Scoring Locatif")
dot.attr(rankdir='LR', fontsize='10', nodesep='1')

# Fonction pour ajouter une entité (table)
def add_entity(name, attributes):
    label = f"<<TABLE BORDER='1' CELLBORDER='1' CELLSPACING='0'>"
    label += f"<TR><TD BGCOLOR='lightblue'><B>{name}</B></TD></TR>"
    for attr in attributes:
        label += f"<TR><TD>{attr}</TD></TR>"
    label += "</TABLE>>"
    dot.node(name, label=label, shape='plaintext')

# ===================== ENTITÉS ===================== #
# USER avec les NOUVEAUX CHAMPS
add_entity("User",[
    "PK id_user",
    "nom", "prenom",
    "email (unique)", "telephone (unique)",
    "date_naissance",
    "type_piece", "numero_piece",
    "date_exp_piece", "img_piece"
])

# CANDIDATURE avec les NOUVEAUX CHAMPS (et référence User)
add_entity("Candidature", [
    "PK id_candidature",
    "revenu_mensuel DECIMAL(12,2)",
    "fonction", "domaine_activite",
    "type_emploi", "type_contrat",
    "charge_mensuelle DECIMAL(12,2)", "dette DECIMAL(12,2)",
    "anciennete_professionnelle INT",
    "nombre_loge_passe INT", "nbr_pers_aCharge INT",
    "FK id_user"
    # NB: les documents sont côté Document via FK id_candidature (1..N)
])

# DOCUMENT avec la FK vers CANDIDATURE
add_entity("Document", [
    "PK id_document",
    "type_document", "document_path",
    "FK id_candidature"
])

add_entity("Locataire", [
    "PK id_user (FK→User)",
    "revenu_mensuel DECIMAL(12,2)",
    "profession", "type_emploi",
    "date_inscription"
])

add_entity("Proprietaire", [
    "PK id_proprietaire",
    "nom", "prenom", "email", "telephone"
])

add_entity("Location", [
    "PK id_location",
    "type_location", "surface",
    "localisation", "loyer_mensuel DECIMAL(12,2)",
    "FK id_proprietaire"
])

add_entity("Contrat", [
    "PK id_contrat",
    "date_debut", "date_fin",
    "montant_loyer DECIMAL(12,2)",
    "FK id_user (Locataire)", "FK id_location"
])

add_entity("Paiement", [
    "PK id_paiement",
    "date_paiement", "montant DECIMAL(12,2)",
    "statut",
    "FK id_contrat"
    # Retard géré en 0..1 via table Retard
])

# Retard enrichi pour les échéances/retards
add_entity("Retard", [
    "PK id_retard",
    "date_echeance", "date_paiement_reel",
    "nb_jours_retard INT", "penalite DECIMAL(12,2)",
    "nouvelle_echeance",
    "statut",
    "UK/FK id_paiement (1-1)"
])

# Appréciation du propriétaire (pour scoring/avis)
add_entity("Appreciation", [
    "PK id_appreciation",
    "note INT (1..5)", "categorie", "fiabilite_paiement BOOL",
    "commentaire", "date_appreciation",
    "FK id_proprietaire", "FK id_locataire", "FK id_contrat"
])

# ===================== RELATIONS ===================== #
dot.edge("User", "Locataire", label="1..1")
dot.edge("User", "Candidature", label="1..n")
dot.edge("Candidature", "Document", label="1..n")

dot.edge("Proprietaire", "Location", label="1..n")
dot.edge("Location", "Contrat", label="1..n")
dot.edge("Locataire", "Contrat", label="1..n")
dot.edge("Contrat", "Paiement", label="1..n")

# Retard 0..1 par paiement (one-to-one à travers une FK unique côté Retard)
dot.edge("Paiement", "Retard", label="0..1")

# Appréciations
dot.edge("Proprietaire", "Appreciation", label="1..n")
dot.edge("Locataire", "Appreciation", label="1..n")
dot.edge("Contrat", "Appreciation", label="1..n")

# Afficher dans Jupyter
dot
```
